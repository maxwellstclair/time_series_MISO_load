# Clean and visualize labels

## Cleaning

```python
from google.colab import files
import pandas as pd
import re
```

Upload the .csv file

```python
uploaded = files.upload()
```

Load the .csv file

```python
filename = next(iter(uploaded))
df = pd.read_csv(filename)
```

Clean by removing all non-letter characters from the Name column.

```python
df['Breed'] = df['Name'].apply(lambda x: re.sub(r'[^A-Za-z]', '', str(x)))
df = df.drop(columns=['Name'])
```

```python
df.head()
```

## EDA

```python
import matplotlib.pyplot as plt
```

```python
label_counts = df['Breed'].value_counts()

plt.figure(figsize=(12, 6))
label_counts.plot(kind='bar')
plt.title('Frequency of Breed')
plt.xlabel('Label')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
breed_counts = df['Breed'].value_counts().reset_index()
breed_counts.columns = ['Breed', 'Count']
```

```python
breed_counts.tail()
```

## Save cleaned data as a new csv file

```python
annotations = 'annotations.csv'
df.to_csv(annotations, index=False)

files.download(annotations)
```
